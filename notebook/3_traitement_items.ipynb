```python
import pandas as pd
import numpy as np


```

```python
df_items_1 = pd.read_csv("../data/item_properties_part1.csv")
df_items_2 = pd.read_csv("../data/item_properties_part2.csv")
df_category = pd.read_csv("../data/category_tree.csv")
```

```python
df_items = pd.concat([df_items_1, df_items_2], axis=0)
df_items.shape
df_items.head()
```

```python
df_items.info()
```

```python
df_items.isna().sum()
df_items.duplicated().sum()
```

```python
# Focus sur les catégories produits

df_items[df_items['property'] == 'categoryid'].head()

df_item_category = df_items[df_items['property'] == 'categoryid']
df_item_category['value'].nunique()

```

```python
# les types  de proprietés des items
proprety_type = df_items['property'].value_counts()

print(proprety_type)
```

```python
df_category.head()
```

```python

# Catégories racines

df_category[df_category['parentid'].isna()]
```

```python
# Jointure items ↔ catégories


df_item_category = df_item_category.copy()
df_item_category['categoryid'] = df_item_category['value'].astype(int)

print(df_item_category.head())
print(df_item_category.shape)   
```

```python
df_item_category = df_item_category.merge(df_category[['categoryid', 'parentid']], on='categoryid', how='left')
print(df_item_category.head())
```

```python
# sauvegarde du dataframe final
df_item_category.to_csv("../data/df_item_clean.csv", index=False)
```
